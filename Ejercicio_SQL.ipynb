```python
cur.execute(" \
  SELECT C.Nombre \
  FROM Libro A LEFT JOIN LibAut B  \
  ON A.Id_Libro = B.Id_Libro\
  LEFT JOIN Autor C \
  ON B.Id_Autor = C.Id_Autor\
  WHERE A.Titulo = 'visual Estudio Net'").fetchall()
```

18. Crear un Procedimiento Almacenado que inserte nuevos Estudiantes


```python
CREATE PROCEDURE SPIestudiante
@id_lector INT,
@ci VARCHAR(9),
@nombre VARCHAR(100),
@apellido VARCHAR(100),
@direccion VARCHAR(100),
@carrera VARCHAR(100),
@edad VARCHAR(9)
AS
BEGIN
    INSERT INTO Estudiante (Id_Lector, CI, Nombre, Apellido, Direccion, Carrera, Edad)
                VALUES(@id_lector, @ci, @nombre, @apellido, @direccion, @carrera, @edad)
END
```

19. Crear un Procedimiento Almacenado que actualice cualquier Libro especificando su cdigo.

```python
CREATE PROCEDURE SPIlibros
@id_libro INT
@titulo VARCHAR(100),
@editorial VARCHAR(100),
@area VARCHAR(100),

AS
BEGIN
UPDATE Libro 
    SET Titulo = @titulo,
        Editorial = @editorial,
        Area = @area
  WHERE Id_Libro = @id_libro
END
```

21. Crear una Funcin (que devuelva una Tabla) que liste los prstamos solicitados por un determinado alumno.

```python
CREATE FUNCTION Prestamos (
    @Nombre VARCHAR(100)
)
RETURNS TABLE
AS
RETURN
    SELECT B,* \
  FROM Libro A LEFT JOIN LibAut B  \
  ON A.Id_Libro = B.Id_Libro\
  LEFT JOIN Estudiante C \
  ON B.Id_Lector = C.Id_Lector\
  WHERE C.Nombre = @Nombre
```
